```python
titlefontsize = 18
axfontsize = 15
axlabelfontsize = 16
figsize=(7,4)
```

```python
plt.figure(figsize=figsize)
x = [0,235,285]
y = [0,0.4,0]
plt.plot(x,y,lw=3)
plt.xlim(0,365)
plt.ylim(0,0.6)
plt.title('Scenario: 1462 - Late, slow melt', fontsize= titlefontsize)
ax = plt.gca()
[tick.label.set_fontsize(axfontsize) for tick in ax.xaxis.get_major_ticks()];
[tick.label.set_fontsize(axfontsize) for tick in ax.yaxis.get_major_ticks()];
plt.xlabel('Day of Water Year',fontsize = axlabelfontsize)
plt.ylabel('Snow Water Equivalent [m]',fontsize = axlabelfontsize)
plt.savefig('./figures/Nr1_scen_1462_late_slow.png',dpi=300,bbox_inches='tight')
```
