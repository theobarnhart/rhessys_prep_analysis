```python
jemez = pd.read_pickle('./data/jemez_raws_lapse.pcl')
```

```python
vcm = pd.read_pickle('./data/vcm_tower_forcings.pcl')
```

```python
datmin = vcm.index.min()
datmax = vcm.index.max()
```

```python
jemez = jemez[datmin:datmax]
```

```python
plt.plot(vcm.Tave,jemez.Vcm_tave,'.')
plt.plot([-30,30],[-30,30],'--k', lw=2)
plt.xlabel('VCM tower Forcings',fontsize=14)
plt.ylabel('JEMEZ lapsed Forcings',fontsize=14)
plt.title('Tave', fontsize=16)
```

```python
plt.plot(vcm.Tmin,jemez.Vcm_tmin,'.')
plt.plot([-30,30],[-30,30],'--k', lw=2)
plt.xlabel('VCM tower Forcings',fontsize=14)
plt.ylabel('JEMEZ lapsed Forcings',fontsize=14)
plt.title('Tmin', fontsize=16)
```

```python
plt.plot(vcm.Tmax,jemez.Vcm_tmax,'.')
plt.plot([-30,30],[-30,30],'--k', lw=2)
plt.xlabel('VCM tower Forcings',fontsize=14)
plt.ylabel('JEMEZ lapsed Forcings',fontsize=14)
plt.title('Tmax', fontsize=16)
```

```python
plt.plot(vcm.P,jemez.Vcm_p/1000.,'.')
plt.plot([0,0.05],[0,0.05],'--k', lw=2)
plt.xlim(0,0.05)
plt.ylim(0,0.05)
plt.xlabel('VCM tower Forcings',fontsize=14)
plt.ylabel('JEMEZ lapsed Forcings',fontsize=14)
plt.title('Precip.', fontsize=16)
```
