```python
import numpy as np
import pandas as pd
import sys, subprocess

sys.path.append('/RHESSys/bin/') # add the rhessys programs to the path if they arent there already
```

# Generate the random distributions of variables

# gw1

```python
n = 50

strt = 1
nd = 300

tmp = np.array(np.random.randint(strt,nd,n))
gw1 = tmp/1000.
```

# gw2

```python
strt = 900
nd = 10

tmp = np.array(np.random.randint(nd,strt,n))
gw2 = tmp/1000.
```

# Create the Simulation

```python
# define the parameters that are not being calibrated
m = 0
ksat = 0
```

```python
prefx = 'cal1_'
command = 'rhessys5.18'
world = '-w /RHESSys/Comocreek/barnhatb/worldfiles/world1_gw_cal_lai'
tec = '-t /RHESSys/Comocreek/barnhatb/tecfiles/tec.gw_cal'
strt = '-st 2001 01 01 01'
nd = '-ed 2012 01 01 05'

for i in xrange(0,n):
    #print str(i+1)+'/'+str(n)
    
    soil = '-s '+str(m)+' '+str(ksat)
    gw = '-gw '+str(gw1[i])+' '+str(gw2[i]) # generate the groundwater argument
    pre = '-pre /RHESSys/Comocreek/barnhatb/out/cal/'+prefx+str(i) # generate the output information 
    
    cmd = command + ' ' + world + ' ' + soil + ' ' + gw + ' ' + '-b' + ' ' + tec + ' ' + pre + ' ' + strt + ' ' + nd
    
    print 'echo '+str(i+1)
    print 
    print cmd
    print
    print 'echo "'+str(i+1)+' - Done"'
    print
    subprocess.call(cmd, shell = True)
    
    #print 'Done, Move to Next Iteration!'

np.save('/RHESSys/Comocreek/barnhatb/out/cal/cal1_gw1.npy', gw1)
np.save('/RHESSys/Comocreek/barnhatb/out/cal/cal1_gw2.npy', gw2)
```
