```python
from mpl_toolkits.basemap import Basemap
shp = '/Volumes/data/como_watershed_wgs84.shp'
```

## Overview Map

```python
m = Basemap(llcrnrlon=-105.75,llcrnrlat=39.95,urcrnrlon=-105.0,
            urcrnrlat=40.15,epsg=3395,
            lon_0=-105.5,lat_0=40.5)

plt.figure(figsize=(10,5))
m.arcgisimage(service='ESRI_Imagery_World_2D', xpixels = 2000,trans=0.5)
m.readshapefile(shp[0:-4],shp.split('/')[-1].split('.')[0], color='r', linewidth=0.5);
m.drawparallels(np.arange(-180, 180,0.1), linewidth=2, color='0.5',zorder=10,labels=[1,0,0,1]);
m.drawmeridians(np.arange(-180, 180,0.1), linewidth=2, color='0.5',zorder=10,labels=[1,0,0,1]);

plt.savefig('./figures/como_overview_map.png',dpi=300,bbox_inches='tight')
```

## Detail Map

```python
m = Basemap(llcrnrlon=-105.59,llcrnrlat=40.025,urcrnrlon=-105.535,
            urcrnrlat=40.06,epsg=3395,
            lon_0=-105.5,lat_0=40.5)

plt.figure(figsize=(8,8))
m.arcgisimage(service='ESRI_Imagery_World_2D', xpixels = 2000,trans=0.5)
m.readshapefile(shp[0:-4],shp.split('/')[-1].split('.')[0], color='r', linewidth=1.);
m.drawparallels(np.arange(35, 45,0.02), linewidth=2, color='0.5',zorder=10,labels=[1,0,0,1]);
m.drawmeridians(np.arange(-110, -100,0.02), linewidth=2, color='0.5',zorder=10,labels=[1,0,0,1]);
plt.savefig('./figures/como_detail_map.png',dpi=300,bbox_inches='tight')
```
