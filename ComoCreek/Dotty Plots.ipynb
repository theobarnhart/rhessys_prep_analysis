```python
import glob
calname = '15'
```

```python
dat = pd.read_csv('/RHESSys/Como/out/cal%s.txt'%(calname),names=['idx','pe','nse','log_nse','rmse'])
```

```python
dat.head()
```

```python
dat.dropna(inplace=True)
```

```python
files = glob.glob('/RHESSys/Como/scripts/cal%s_*.txt'%(calname))
```

```python
df = pd.DataFrame()
keys = []
for fl in files:
    key = fl.split('/')[-1].split('_')[-1].split('.')[0]
    
    df[key] = np.loadtxt(fl)
    keys.append(key)
df['idx'] = np.arange(0,len(df))
```

```python
df = pd.merge(left = dat,right = df, on='idx',how='inner')
```
