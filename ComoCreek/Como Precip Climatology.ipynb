## Generate a Plot of Precipitation Climatology to Investigate Long-Term Storage Changes

```python
dat = pd.read_pickle('./data/como_gridded_summary.pcl') # load the data
dat = dat.groupby('wateryear').sum() # summarize by water year
dat['Panom'] = dat.P - dat.P.mean() # subtract the mean P from P
```

```python
dat.Panom.plot.bar(figsize=(15,5),facecolor='0.5')
plt.axhline(0,color='k')
plt.ylim(-0.55,0.55)
plt.xlabel('Wateryear', fontsize = 16)
plt.ylabel('Precipitation Anomoly\n $(P - \overline{P})$', fontsize=16)
plt.axvspan(16.5,21.5,color='0.7',alpha=0.2)
#plt.savefig('./figures/como_precip_climatology.png',dpi=300,bbox_inches='tight')
```
