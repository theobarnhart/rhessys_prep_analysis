```python
def parse_date(df):
    return '%s-%s-%s'%(int(df.year),int(df.month),int(df.day))
```

```python
def read_file(fl):
    dat = pd.read_table(fl,sep=' ')
    dat['datetime'] = dat.apply(parse_date,axis=1)
    dat.index = pd.DatetimeIndex(dat.datetime)
    dat['ET'] = dat.trans + dat.evap
    return dat
```
