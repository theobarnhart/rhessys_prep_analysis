```python
dat = dat.loc['1989-10-01':'2013-09-30',:]
```

```python
np.sum(np.isnan(dat.ATmin))
```

```python
del dat['date']
del dat['swe']
del dat['precip_accum']
```

```python
reps = 25
nd = pd.datetime(2250,9,30)

ATmax = np.tile(dat.ATmax.as_matrix(),reps)
ATmin = np.tile(dat.ATmin.as_matrix(),reps)
precip = np.tile(dat.precip.as_matrix(),reps)

td = pd.datetools.timedelta(len(ATmax))

print 'Start Date:',nd-td
print 'End Date:',nd

strt = nd-td

strtlab = str(strt.year)+' '+str(strt.month)+' '+str(strt.day)
```

```python
np.savetxt('/RHESSys/ComoCreek_new/clim/niwot_snotel_su.tmax',ATmax,fmt='%10.3f',header=strtlab,comments='')
np.savetxt('/RHESSys/ComoCreek_new/clim/niwot_snotel_su.tmin',ATmin,fmt='%10.3f',header=strtlab,comments='')
np.savetxt('/RHESSys/ComoCreek_new/clim/niwot_snotel_su.rain',precip,fmt='%10.3f',header=strtlab,comments='')
```
