```python
def classify(value, kclass):
    for i in range(1,len(kclass)): # loop through each class
        if value < kclass[i]:
            return i
        elif value > kclass[i]:
            continue
```

```python
for i in xrange(0, len(dem2)):
    dem_class[i] = classify(dem2[i], kclass)
```

### Reinsert the classified DEM into the DEM shape

```python
m,n = np.shape(dem)

dem3 = np.reshape(dem,[(m*n),1]) # copy the DEM
```

```python
np.shape(dem_class)
```

```python
for i in xrange(0,len(dem3[np.isnan(dem3)==False])):
    dem3[np.isnan(dem3) == False][i] = dem_class[i]
```

```python
dem3 = np.reshape(dem3,[m,n])
```
